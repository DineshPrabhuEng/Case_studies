### Importing secondary data frame

```python
df_country=pd.read_excel('/content/drive/MyDrive/Zomato/Country-Code.xlsx')
```

```python
df_country.head()
```

### merging the both data frames

```python
final_df=df.merge(df_country,on='Country Code',how='left')
```

```python
final_df.dtypes
```

```python
#Unique values in all the columns
final_df.Country.value_counts()
```

**Observation**:Zomato business is widely spread in india

```python
country_name=final_df.Country.value_counts().index #returns unique values (sorted)
# final_df.Country.unique() #returns unique values
```

```python
values=final_df.Country.value_counts().values
```

**PIE Chart**

```python
plt.figure(figsize=(10,6))
plt.title('Top 5 Countries')
plt.pie(values[:4],labels=country_name[:4],autopct='%1.2f%%')
plt.show()
```

**Observation**: inida contributes 93.78% of Zomato Business

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

**Rating Classification**

1.   Rating from 1.8 to 2.4---Poor
2.   Rating from 2.5 to 3.4---Average
3.   Rating from 3.5 to 3.9---Good
4.   Rating from 4.0 to 4.4--Very Good
5.   Rating from 4.5 to 4.9 ---Excellent



```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=rating)
```

```python

sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=rating,palette=['white','red','orange','yellow','green','green'])
```

**Maximum number of ratings are between 3.0 to 3.6**

```python
r1=pd.DataFrame(rating.groupby(['Rating text'])['Rating Count'].sum())
#why only one column is showing i want to see a name for aggregated column
```
